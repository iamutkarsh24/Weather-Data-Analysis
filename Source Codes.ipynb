```python
# Importing of the libraries
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
# Load/Read the dataset
df = pd.read_csv('Weather_Dataset.csv')
```

```python
# To check the null values
df.isna().sum()
```

```python
# Drop all the duplicate values. (if any)
df.drop_duplicates(inplace=True)
```

```python
# Conversion into Date/Time
df['Date/Time'] = pd.to_datetime(df['Date/Time'])
```

```python
# Seperate columns
df['Date'] = df['Date/Time'].dt.date
df['Time'] = df['Date/Time'].dt.time
```

### Average visibility 

```python
df.groupby(df['Weather'],as_index=False)['Visibility_km'].mean().sort_values(by='Visibility_km',ascending=False)
```

### Weather that contains 'Snow'

```python
df[df['Weather'].str.contains('Snow')]
```

### Weather that contains 'Fog'

```python
df[df['Weather'].str.contains('Fog')]
```

### Data where the wind speed>25 and visibility =25

```python
df[(df['Wind Speed_km/h']>25) & (df['Visibility_km']==25)]
```

### Either Weather is 'Clear' or Visibility is greater than 20

```python
df[(df['Weather']=='Clear') | (df['Visibility_km']>20)]
```

### All the means values in respect to weather column

```python
numerical_cols =['Temp_C','Dew Point Temp_C','Rel Hum_%','Wind Speed_km/h','Visibility_km','Press_kPa']
df.groupby(df['Weather'])[numerical_cols].mean()
```

### All the max values in respect to Weather column

```python
numerical_cols =['Temp_C','Dew Point Temp_C','Rel Hum_%','Wind Speed_km/h','Visibility_km','Press_kPa']
df.groupby(df['Weather'])[numerical_cols].max()
```

### All the min values in respect to Weather column

```python
numerical_cols =['Temp_C','Dew Point Temp_C','Rel Hum_%','Wind Speed_km/h','Visibility_km','Press_kPa']
df.groupby(df['Weather'])[numerical_cols].min()
```

### Average visibility each hours

```python
df.groupby(df['Time'])[['Visibility_km']].mean()
```
